Neste notebook serão feitas a exploração dos dados, com exibição de gráficos relacionando as colunas e suas
respectivas interpretações.

```python
import matplotlib.pyplot as plt
import seaborn as sbn
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
pd_filmes = pd.read_json("filmes.json", )
```

```python
pd_filmes.index = pd_filmes.index.order()
```

### Filmes mais bem avaliados por ano pelo público

```python
max_rating_per_year = pd_filmes.groupby(['year']).max()
```

```python
max_rating_per_year[['pt_title','rating']]
```

```python
mean_rating_per_year = pd_filmes.groupby(['year']).mean()
```

### Avaliacão média dos filmes por ano dada pelo público (rating)

```python
mean_rating_per_year[['rating']]
```

###### Observe que a nota dada foi descrescente ao longo dos anos

```python
mean_rating_per_year[['rating']].plot(figsize=(20,10))

plt.xlim(1912, 2016)

plt.show()
```

### Os filmes mais votados por ano

```python
max_rating_per_year[['pt_title','votes']]
```

```python
max_rating_per_year[['votes']].plot(figsize=(20,10))
plt.xlim(1912, 2016)
plt.show()
```

```python
pd_filmes.plot(kind = 'scatter', x = 'year', y = 'rating', figsize=(20,10))

plt.xlim(1912, 2015)

plt.show()
```

### Relação entre duração do filme e avaliação dada pelo público

```python
pd_filmes.plot(kind = 'scatter', x = 'running_time', y = 'rating', figsize=(20,10))
plt.show()
```

Há poucos filmes com menos de 50 minutos ou com mais de 200 minutos de duração.  Do gráfico, também podemos notar que a avaliação dada aos filmes com mais de 200 minutos é, em geral, bem alta.

### Relação entre a quantidade de votos e a avaliação dada pelo público

```python
pd_filmes.plot(kind = 'scatter', x = 'votes', y = 'rating', figsize=(20,10))
xmin, xmax = plt.xlim()
plt.xlim(0, 140000)
plt.show()
```

### Quantidade de filmes por ano

```python
movies_per_year = pd_filmes.groupby('year').count()
```

```python
plt.figure(figsize=(20,10))

movies_per_year.plot(kind='bar', y = 'pt_title')

plt.show()
```

### Quantidade de filmes por avaliação

```python
movies_rating = pd_filmes.groupby('rating').count()

plt.figure(figsize=(20,10))

movies_rating.plot(kind='bar', y = 'pt_title')

plt.show()
```

### Quantidade de filmes por duração

```python
movies_running = pd_filmes[pd_filmes.running_time>0].groupby('running_time').count()

plt.figure(figsize=(20,10))

movies_running.plot(kind='bar', y = 'pt_title')

plt.show()
```

```python
pd_filmes.plot(kind = 'scatter', x = 'votes', y = 'running_time', figsize=(20,10))

plt.xlim(2000, 1400000)
plt.ylim(0,500)
plt.show()
```
