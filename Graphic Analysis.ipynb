Dicionário relacionando ator(chave) a todos os filmes em que ele atuou (lista) 

```python
actors = {}
for item in pd_filmes.values:
    for actor in item[10]:
        if actor not in actors.keys():
            actors[actor] =  []
            actors[actor].append(item[7])
        else:
            actors[actor].append(item[7])
```

```python
print actors['Tim Robbins']
```

```python
print len(actors) 
```

há 9883 atores envolvidos nos filmes analisados

#### Atores que atuaram em mais filmes

```python
values = []
for actor in actors.keys():
    pair = (actor.encode('utf-8'), len(actors[actor]))
    values.append(pair)
dtype = [('name', 'S30'), ('Qt_movies', int)]

mv = np.array(values, dtype = dtype)
```

```python
famous_actors = np.sort(mv, order='Qt_movies', kind = 'mergesort')
```

```python
for pair in famous_actors[::-1]:
    print pair[0], '=>', pair[1]
```
